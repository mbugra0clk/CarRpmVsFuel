RÄ°DGE ROSSO  Ã‡ok deÄŸiÅŸkenli regresyon verilerini analiz etmede kullanÄ±lÄ±r. AmaÃ§ hata kareler toplamÄ±nÄ± minimize eden katsayÄ±larÄ±, bu katsayÄ±lara bir ceza uygulayarak bulmaktÄ±r.
ridge rosso fonk. 50 adet Ã¼zeri data ile Ã§alÄ±ÅŸÄ±r.

```python
import pandas as pd
df = pd.read_csv("Car_price_cleaned.csv")
df= df.select_dtypes(include = ['float64', 'int64'])
X = df
y = df["peakrpm"]

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha = 0.1).fit(X_train, y_train)
```

```python
ridge_model
```

```python
ridge_model.coef_
```

```python
10**np.linspace(10,-2,100)*0.5 
```

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 

ridge_model = Ridge()
katsayilar = []

for i in lambdalar:
    ridge_model.set_params(alpha = i)
    ridge_model.fit(X_train, y_train) 
    katsayilar.append(ridge_model.coef_) 
    

    
ax = plt.gca()
ax.plot(lambdalar, katsayilar) 
ax.set_xscale('log') 

plt.xlabel('Lambda(Alpha) DeÄŸerleri')
plt.ylabel('KatsayÄ±lar/AÄŸÄ±rlÄ±klar')
plt.title('DÃ¼zenlileÅŸtirmenin Bir Fonksiyonu Olarak Ridge KatsayÄ±larÄ±');
```

```python
y_pred = ridge_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 
```

```python
lambdalar[0:5]
```

```python
from sklearn.linear_model import RidgeCV
ridge_cv = RidgeCV(alphas = lambdalar, 
                   scoring = "neg_mean_squared_error",
                   normalize = False)
```

```python
ridge_cv.fit(X_train, y_train)
```

```python
ridge_cv.alpha_
```

```python
ridge_tuned = Ridge(alpha = ridge_cv.alpha_, 
                   normalize = True).fit(X_train,y_train)
```

```python
np.sqrt(mean_squared_error(y_test, ridge_tuned.predict(X_test)))
```
